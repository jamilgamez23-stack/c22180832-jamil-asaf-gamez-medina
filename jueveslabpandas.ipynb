```python
import pandas as pd
from google.colab import files

print("Por favor, sube tu archivo 'vgsales.csv' cuando se te pida.")
uploaded = files.upload()


file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

print("\n--- Primeras 5 filas del DataFrame ---")
print(df.head())

print("\n--- Información general del DataFrame (tipos de datos y valores nulos) ---")
print(df.info())


print("\n--- Juegos de 'Wii' en el género 'Sports' ---")
juegos_wii_sports = df[(df['Platform'] == 'Wii') & (df['Genre'] == 'Sports')]
print(juegos_wii_sports.head())

print("\n--- Top 5 juegos por ventas globales ---")
df_top_ventas = df.sort_values(by='Global_Sales', ascending=False)
print(df_top_ventas.head())

print("\n--- Ventas globales totales por plataforma ---")
ventas_por_plataforma = df.groupby('Platform')['Global_Sales'].sum().reset_index()

ventas_por_plataforma_ordenadas = ventas_por_plataforma.sort_values(by='Global_Sales', ascending=False)
print(ventas_por_plataforma_ordenadas.head(10))


print("\n--- DataFrame con nueva columna 'NA_EU_Sales' ---")
df['NA_EU_Sales'] = df['NA_Sales'] + df['EU_Sales']
print(df.head())

# Puedes guardar los resultados de tu análisis en un nuevo archivo CSV si lo deseas.
# df.to_csv('resultados_analisis.csv', index=False)
# print("\nResultados guardados en 'resultados_analisis.csv'")
```
